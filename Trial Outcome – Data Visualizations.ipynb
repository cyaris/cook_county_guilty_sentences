```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_colwidth', 1000)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df = pd.read_pickle('./Pickles/df_modeling.pkl')
```

```python
plt.figure(figsize = (12, 8))
sns.set(font_scale=1.8, style = 'white')

ax = sns.countplot(x = df['found_guilty'], color = '#1D9A78')
# ax.set_title('_______')
ax.set_ylabel('')
ax.set_xlabel('')
ax.set_xticklabels(['Pled Guilty', 'Found Guilty at Trial'])

plt.tight_layout()
```
